```python
1. **a) Mean**  
The mean is the average of a set of numbers, calculated by dividing the sum of all values by the number of values.

**b) Median**  
The median is the middle most in a sorted listed of numbers.

**c) Mode**  
The mode is the value that appears most frequently in a dataset.
```

```python
import numpy as np
from scipy import stats

data = [2, 4, 4, 6, 8, 10]

mean = np.mean(data)
median = np.median(data)
mode = stats.mode(data, keepdims=True).mode[0]

mean, median, mode
```

```python
2. **Variance** measures the average of the squared differences from the Mean.  
**Standard Deviation** is the square root of the variance and represents data spread in the same unit as the data.
```

```python
sample_data = [5, 7, 3, 7, 10, 15]

variance = np.var(sample_data)
std_deviation = np.std(sample_data)

variance, std_deviation
```

```python
4. It states that the distribution of the sample mean approximates a normal distribution as the sample size gets larger, regardless of the population's distribution.

**Demonstration using Python:**
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

population = np.random.exponential(scale=2, size=10000)

sample_means = [np.mean(np.random.choice(population, size=30)) for _ in range(1000)]

sns.histplot(sample_means, kde=True)
plt.title("Distribution of Sample Means (Central Limit Theorem)")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.show()
```

```python
#Data Visualization using Seaborn & Matplotlib
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

tips = sns.load_dataset("tips")
iris = sns.load_dataset("iris")
penguins = sns.load_dataset("penguins")
```

```python
1. **Plot a histogram of the total_bill column from the tips dataset
python**
 
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(tips['total_bill'], bins=20, kde=True)
plt.title("Histogram of Total Bill")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.show()
```

```python
2. **Create a scatterplot of total_bill vs tip and add hue for the sex column**
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="sex")
plt.title("Scatterplot of Total Bill vs Tip by Gender")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
3. **Draw a boxplot of total_bill for each day of the week using the tips dataset**
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=tips, x="day", y="total_bill")
plt.title("Boxplot of Total Bill by Day")
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.show()
```

```python
4. **Create a countplot showing the frequency of each day in the tips dataset**
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=tips, x="day")
plt.title("Count of Records by Day")
plt.xlabel("Day")
plt.ylabel("Count")
plt.show()
```

```python
5. **Plot a barplot showing the average tip per day**
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=tips, x="day", y="tip", estimator=np.mean)
plt.title("Average Tip per Day")
plt.xlabel("Day")
plt.ylabel("Average Tip")
plt.show()
```

```python
6. **Create a KDE plot of petal_length from the iris dataset
python**
```

```python
plt.figure(figsize=(8, 5))
sns.kdeplot(data=iris['petal_length'], fill=True)  # updated from shade=True to fill=True
plt.title("KDE Plot of Petal_Length (Iris Dataset)")
plt.xlabel("Petal_Length")
plt.show()
```

```python
7. **Use a pairplot on the iris dataset to visualize relationships between all features**
```

```python
sns.pairplot(data=iris, hue="species")
plt.suptitle("Pairplot of Iris Dataset", y=1.02)
plt.show()
```

```python
8. **Plot a violinplot showing the distribution of tip amounts by smoker status**
```

```python
plt.figure(figsize=(8, 5))
sns.violinplot(data=tips, x="smoker", y="tip")
plt.title("Violinplot of Tip by Smoker Status")
plt.xlabel("Smoker")
plt.ylabel("Tip")
plt.show()
```

```python
9.**Detect outliers in the total_bill column using the IQR method and display them**
```

```python
Q1 = tips['total_bill'].quantile(0.25)
Q3 = tips['total_bill'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = tips[(tips['total_bill'] < lower_bound) | (tips['total_bill'] > upper_bound)]
outliers
```

```python
import numpy as np
```

```python
data = [25, 28, 30, 29, 27, 45, 26, 31, 120, 32, 30, 29, 28, 27, 26, 25, 90]
```

**Calculate Q1 (25th percentile)**

```python
Q1 = np.percentile(data, 25)
print(f"Q1 (25th percentile): {Q1}")
```

**Calculate Q3 (75th percentile)**

```python
Q3 = np.percentile(data, 75)
print(f"Q3 (75th percentile): {Q3}")
```

**Calculate the IQR**

```python
IQR = Q3 - Q1
print(f"IQR (Q3 - Q1): {IQR}")
```

**Calculate Lower and Upper Bounds**

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
print(f"Lower Bound: {lower_bound}")
print(f"Upper Bound: {upper_bound}")
```

**Identify Outliers**

```python
outliers = [x for x in data if x < lower_bound or x > upper_bound]
print(f"Outliers: {outliers}")
```
